```python
# reading images as numpy array
# pip install imageip
import imageio
import numpy as np
import matplotlib.pyplot as plt

pic = imageio.imread("demo_2.jpg")

print(pic.shape)
```

```python
print(pic[0,0,:3])
```

```python
plt.figure(figsize=(12,10))
plt.imshow(pic)
plt.show()
```

```python
res = np.dot([101,131,61],[0.299,0.587,0.114])
print(res)
```

```python
g = lambda x: np.dot(x[:,:,:3],[0.299,0.587,0.114])

gray = g(pic)

plt.figure(figsize=(12,10))
plt.imshow(gray,cmap=plt.get_cmap(name='gray'))
plt.show()
```
